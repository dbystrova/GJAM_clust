# Compute the prior distribution with random Gamma

```julia
using Distributions 
Alpha = rand(Gamma(1.939519, 1/0.3114096),1000)
mean(Alpha)
DF_DPM = map(x ->Pkn_Dirichlet_Mult.(1:112,112,112,x),Alpha)
```

```julia
DF_DPM_16= hcat(DF_DPM...)'
Ar =mean(DF_DPM_16,dims=1)
```

```julia
L= Array{Float64}(undef, 112) 
for l in (1:112)
  L[l]= Ar[1,l]
end

L |> ar -> map(*, ar, 1:112) |> sum
```

```julia
R"library(tidyverse)
library(latex2exp)"

R"p = ggplot(data.frame(x = 1:112,
                        Pkn_PYM_05 = $(Pkn_PYM.(1:112,112,112,0.48,0.5)),
                        Pkn_DP_gjam = $(Pkn_Dirichlet_Mult.(1:112,112,112,112)),
                        Pkn_DP1_fixed = $(Pkn_Dirichlet_Mult.(1:112,112,112,6.23)),
                        Pkn_DP1 = $L
                    ) %>%
            gather(Process_type, density, Pkn_PYM_05:Pkn_DP1),
    aes(x=x, y = density, colour = Process_type)) +
geom_line() +
ggthemes::scale_colour_ptol() +
theme_minimal()"
R"p"
R"png('Priors_k_16.png')
plot(p)
dev.off()"
```

```julia
using Plots
plot(1:112,L)
```

```julia
R" 
DPM_prior_16= $L
save(DPM_prior_16,file ='DPM_prior_16.Rdata')
"
```
